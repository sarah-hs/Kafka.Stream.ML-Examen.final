```python
from pyspark.streaming import StreamingContext
from time import sleep
from pyspark import SparkContext
from pyspark.sql import SparkSession
import json

from pyspark.streaming import StreamingListener
from lightning import Lightning
import requests
import numpy as np
```

```python
#init lightning-viz server #lgn = Lightning(local=True)
from numpy import random
lgn = Lightning(host="http://localhost:3000")
lgn.create_session("Rojo")
 
series = random.randn(3,1)
viz = lgn.linestreaming(series)
print(series)
```

```python
sc = SparkContext(appName="ExamenFinal")
ssc = StreamingContext(sc, 2)
ssc.checkpoint('/tmp')
lines = ssc.socketTextStream('localhost', 9009)

jsons = lines.map(lambda x: json.loads(x))

lugar1 = jsons.map(lambda x: (x['lugar1']['location']['name'], x['lugar1']['current']))
lugar2 = jsons.map(lambda x: (x['lugar2']['location']['name'], x['lugar2']['current']))
lugar3 = jsons.map(lambda x: (x['lugar3']['location']['name'], x['lugar3']['current']))
lugar1.pprint()
lugar2.pprint()
lugar3.pprint()

cols = 'wind_degree'

def measures(data, col):
    dataList = []
    dataList.append([data['lugar1']['current'][col]])
    dataList.append([data['lugar2']['current'][col]])
    dataList.append([data['lugar3']['current'][col]])
    return dataList
    
def sendRecord(record):
    viz.append(np.array(record))

check1 = jsons.map(lambda x: measures(x, cols))
check1.foreachRDD(lambda rdd: rdd.foreach(sendRecord))
#counts = pairs.reduceByKey(lambda a, b: a + b)

#ssc.addStreamingListener(SparkMonitoringListener())
ssc.start()
#sleep(5)
#ssc.stop(stopSparkContext=False, stopGraceFully=True)
ssc.awaitTermination()
```
